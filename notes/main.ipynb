```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import os
```

```python
with open('files/selected.log') as file:
    train_lines = [line.rstrip() for line in file]
train_lines = [line.split() for line in train_lines if '] Epoch: [' in line]
train_lines = np.array(train_lines)
```

```python
arr = np.column_stack(
    ([
        [row[1:-1].split('][')[0].split('/')[0],
         row[1:-1].split('][')[1].split('/')[0]] for row in train_lines[:,8]]
       ,train_lines[:,17]
       ,[x[:-1] for x in train_lines[:,19]]
)).astype(float)

new_arr = []
for i in range(1,162):
    new_arr.append([i] + list(arr[arr[:,0]==i].mean(axis = 0)[2:]))
```

```python
df = pd.DataFrame(new_arr, columns = ['Epoch','Loss','Acc'])
```

```python
df
```

```python
plt.plot(df['Epoch'], df['Loss'])
plt.title('Epoch vs Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
plt.plot(df['Epoch'], df['Acc'])
plt.title('Epoch vs Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Acc')
plt.show()
```

```python
with open('files/selected.log') as file:
    train_lines = [line.rstrip() for line in file]
train_lines = [line[-len('84.9119.'):-1] for line in train_lines[:-1] if 'avgAcc ' in line]
train_lines = np.array(train_lines).reshape((len(train_lines)//2,2)).astype(float)
df = pd.DataFrame(np.column_stack((np.arange(1,len(train_lines)+1),
                                   np.array(train_lines))), columns=['Epoch','Train Acc','Val Acc'])
```

```python
plt.plot(df['Epoch'], df['Train Acc'], label = 'Train')
plt.plot(df['Epoch'], df['Val Acc'], label = 'Val')
plt.title('Epoch vs Training Acc vs Val Acc')
plt.xlabel('Epoch')
plt.ylabel('Acc')
plt.legend()
plt.show()
```
