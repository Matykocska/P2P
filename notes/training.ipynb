```python
import os
import numpy as np
import pandas as pd
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
```

```python
exp_root = '../Exp'
val_acc          = 'avgAcc_validate.txt'
val_loss         = 'avgLoss_validate.txt'
train_epoch_acc  = 'comu_epoch_acc.txt'
train_epoch_loss = 'comu_epoch_loss.txt'
train_iter_acc   = 'comu_Iter_acc.txt'
train_iter_loss  = 'comu_Iter_loss.txt'

shapenetpart_exp = os.path.join(exp_root, 'shapenetpart')

# Training model from scratch
shapenetpart_exp5 = os.path.join(shapenetpart_exp, 'p2p_ResNet-18_exp5')

# Transfer learning for model trained on ModelNet40
# Image model parameters are frozen
shapenetpart_exp4 = os.path.join(shapenetpart_exp, 'p2p_ResNet-18_exp4')

# Transfer learning for model trained on ModelNet40
# Encoder and image model parameters are frozen 
shapenetpart_exp3 = os.path.join(shapenetpart_exp, 'p2p_ResNet-18_exp3')

shapenetcore_exp = os.path.join(exp_root, 'shapenetcore')

# Training model from scratch
shapenetcore_exp1 = os.path.join(shapenetcore_exp, 'p2p_ResNet-18_exp1')

# Transfer learning for model trained on ModelNet40
# Image model parameters are frozen
shapenetcore_exp2 = os.path.join(shapenetcore_exp, 'p2p_ResNet-18_exp2')

# Transfer learning for model trained on ModelNet40
# Encoder and image model parameters are frozen 
shapenetcore_exp3 = os.path.join(shapenetcore_exp, 'p2p_ResNet-18_exp3')
```

## ShapeNetPart results

```python
snp_exp5_data =  np.column_stack((np.genfromtxt(os.path.join(shapenetpart_exp5, train_epoch_acc)),
                                  np.genfromtxt(os.path.join(shapenetpart_exp5, train_epoch_loss)),
                                  np.genfromtxt(os.path.join(shapenetpart_exp5, val_acc)),
                                  np.genfromtxt(os.path.join(shapenetpart_exp5, val_loss))))
snp_exp5_data = np.column_stack((np.arange(len(snp_exp5_data)), snp_exp5_data))
snp_exp5_df = pd.DataFrame(snp_exp5_data, columns=['Epoch', 'Train Acc', 'Train Loss', 'Val Acc', 'Val Loss'])

snp_exp4_data =  np.column_stack((np.genfromtxt(os.path.join(shapenetpart_exp4, train_epoch_acc)),
                                  np.genfromtxt(os.path.join(shapenetpart_exp4, train_epoch_loss)),
                                  np.genfromtxt(os.path.join(shapenetpart_exp4, val_acc)),
                                  np.genfromtxt(os.path.join(shapenetpart_exp4, val_loss))))
snp_exp4_data = np.column_stack((np.arange(len(snp_exp4_data)), snp_exp4_data))
snp_exp4_df = pd.DataFrame(snp_exp4_data, columns=['Epoch', 'Train Acc', 'Train Loss', 'Val Acc', 'Val Loss'])

snp_exp3_data =  np.column_stack((np.genfromtxt(os.path.join(shapenetpart_exp3, train_epoch_acc)),
                                  np.genfromtxt(os.path.join(shapenetpart_exp3, train_epoch_loss)),
                                  np.genfromtxt(os.path.join(shapenetpart_exp3, val_acc)),
                                  np.genfromtxt(os.path.join(shapenetpart_exp3, val_loss))))
snp_exp3_data = np.column_stack((np.arange(len(snp_exp3_data)), snp_exp3_data))
snp_exp3_df = pd.DataFrame(snp_exp3_data, columns=['Epoch', 'Train Acc', 'Train Loss', 'Val Acc', 'Val Loss'])
```

```python
plt.plot(snp_exp5_df['Epoch'], snp_exp5_df['Train Acc'], label = "Training Accuracy")
plt.plot(snp_exp5_df['Epoch'], snp_exp5_df['Val Acc'], label = "Validation Accuracy")
plt.title('Training vs Validation Accuracy ({0})'.format("Exp 5"))
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(70,100)
plt.show()
```

```python
fig = plt.figure(figsize=(30, 8))
plt.title("ShapeNetPart Training and Validation Accuracy in different experiments", fontsize=20, y=1.1)
plt.axis('off')

fig.add_subplot(1, 3, 1)
plt.plot(snp_exp5_df['Epoch'], snp_exp5_df['Train Acc'], label = "Training Accuracy")
plt.plot(snp_exp5_df['Epoch'], snp_exp5_df['Val Acc'], label = "Validation Accuracy")
plt.title('Training model from scratch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(70,100)
plt.legend(loc='lower right')

fig.add_subplot(1, 3, 2)
plt.plot(snp_exp4_df['Epoch'], snp_exp4_df['Train Acc'], label = "Training Accuracy")
plt.plot(snp_exp4_df['Epoch'], snp_exp4_df['Val Acc'], label = "Validation Accuracy")
plt.title('Transfer learning with image model parameters frozen')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(70,100)
plt.legend(loc='lower right')

fig.add_subplot(1, 3, 3)
plt.plot(snp_exp3_df['Epoch'], snp_exp3_df['Train Acc'], label = "Training Accuracy")
plt.plot(snp_exp3_df['Epoch'], snp_exp3_df['Val Acc'], label = "Validation Accuracy")
plt.title('Transfer learning with encoder and image model parameters frozen')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(70,100)
plt.legend(loc='lower right')

plt.show()
```

```python
fig = plt.figure(figsize=(30, 8))
plt.title("ShapeNetPart Training and Validation Loss in different experiments", fontsize=20, y=1.1)
plt.axis('off')

fig.add_subplot(1, 3, 1)
plt.plot(snp_exp5_df['Epoch'], snp_exp5_df['Train Loss'], label = "Training Loss")
plt.plot(snp_exp5_df['Epoch'], snp_exp5_df['Val Loss'], label = "Validation Loss")
plt.title('Training model from scratch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(1.0, 1.7)
plt.legend(loc='lower right')

fig.add_subplot(1, 3, 2)
plt.plot(snp_exp4_df['Epoch'], snp_exp4_df['Train Loss'], label = "Training Loss")
plt.plot(snp_exp4_df['Epoch'], snp_exp4_df['Val Loss'], label = "Validation Loss")
plt.title('Transfer learning with image model parameters frozen')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(1.0, 1.7)
plt.legend(loc='lower right')

fig.add_subplot(1, 3, 3)
plt.plot(snp_exp3_df['Epoch'], snp_exp3_df['Train Loss'], label = "Training Loss")
plt.plot(snp_exp3_df['Epoch'], snp_exp3_df['Val Loss'], label = "Validation Loss")
plt.title('Transfer learning with encoder and image model parameters frozen')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(1.0, 1.7)
plt.legend(loc='lower right')

plt.show()
```
